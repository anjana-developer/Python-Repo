<h4>Inspecting duplicates<h4>

```python
df.nunique()
```

```python
#though there are 690 rows ! patient id has some duplicate values--> IDENTIFY THEM!!

df[df.duplicated(subset = 'patient_id', keep =False)].sort_values('patient_id')
```

```python
#group each patient id which has been repeated


repeated_patients = df.groupby(by = 'patient_id').size().sort_values(ascending =False)
repeated_patients
```

```python
#Filtering data
#adding a tilde the pandas boolean series is reversed and thus the resulting data frame is of those that do NOT repeat more than twice.
#to identify id s which are nt repeated more tha twice

filtered_patients = repeated_patients[repeated_patients > 2].to_frame().reset_index()
filtered_df = df[~df.patient_id.isin(filtered_patients.patient_id)]
filtered_df

```

```python
#RE-SHAPING DATA

reshaped_data = df[['patient_id','doctor_name']]

reshaped_data
```

```python
#Adding an extra column

reshaped_data['doctor_count']=1
reshaped_data
```

```python
#HOT-ENCODING

doctors_one_hot_encoded = pd.pivot_table( reshaped_data,
                                  index = reshaped_data.index, 
                                  columns = ['doctor_name'], 
                                   values = ['doctor_count'] )
```

```python
doctors_one_hot_encoded = doctors_one_hot_encoded.fillna(0)
```

```python
doctors_one_hot_encoded.columns = doctors_one_hot_encoded.columns.droplevel()
```

```python
doctors_one_hot_encoded
```

```python
#merge
#syntax
#leftJoin_df = pd.merge(df1, df2, on ='col_name', how='left')

combined_df = pd.merge(df, doctors_one_hot_encoded, left_index = True,right_index =True)
```

```python
combined_df.drop(columns=['doctor_count'])
combined_df
                      
```

```python
#ROW WISE OPERATIONS

def normalorabnormal(x):
    if ((x['cell_size_uniformity'] > 5) &      (x['cell_shape_uniformity'] > 5)):
       print('normal')
    else:
       print('abnormal')
        
#to return the fun to the daataframe

combined_df['normalorabnormal'] = combined_df.apply(lambda x: normalorabnormal(x), axis=1)
combined_df
```
